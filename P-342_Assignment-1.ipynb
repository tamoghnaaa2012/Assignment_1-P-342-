```python
#Question 2:(factorial n! say for n=10 or 15, check and trap negative integers, say for -5!)


n = int(input("Enter the number:"))      #will find the factorial of this number

result = 1                             #Initialisation
for i in range(n,1,-1):                #range(start, stop, step)
    result = result*i
if n< 0:
    print("Factorial cannot be found for negative numbers")        # if input number is negative then return an error message
elif n== 0:
    print("Factorial of 0 is 1")                                   # elif the input number is 0 then display 1 as output
else:
    print("Factorial of", n, "is: ", result)
```
